# Query 1

```python
# How many food providers and receivers are there in each city?

query = """
SELECT 
    City,
    COUNT(DISTINCT Provider_ID) AS Total_Providers,
    COUNT(DISTINCT Receiver_ID) AS Total_Receivers
FROM (
    SELECT City, Provider_ID, NULL AS Receiver_ID FROM Providers
    UNION ALL
    SELECT City, NULL AS Provider_ID, Receiver_ID FROM Receivers
) AS combined
GROUP BY City
ORDER BY City;
"""

result_df = pd.read_sql(query, conn)
result_df
```

# Query 2

```python
# Which type of food provider (restaurant, grocery store, etc.) contributes the most food?

query = """
SELECT TOP 1 
    p.Type AS Provider_Type,
    SUM(f.Quantity) AS Total_Quantity
FROM Providers p
JOIN Food_Listings_Dataset f
    ON p.Provider_ID = f.Provider_ID
GROUP BY p.Type
ORDER BY Total_Quantity DESC;
"""

result_df = pd.read_sql(query, conn)
result_df
```

# Query 3

```python
# What is the contact information of food providers in a specific city?

city_name = 'Adambury'

query = f"""
SELECT 
    Name,
    Type,
    Address,
    City,
    Contact
FROM Providers
WHERE City = '{city_name}'
ORDER BY Name
OFFSET 0 ROWS FETCH NEXT 5 ROWS ONLY;
"""

result_df = pd.read_sql(query, conn)
result_df
```

# Query 4

```python
# Which receivers have claimed the most food?

query = """
SELECT 
    r.Receiver_ID,
    r.Name AS Receiver_Name,
    COUNT(c.Claim_ID) AS Total_Claims
FROM Claims c
JOIN Receivers r 
    ON c.Receiver_ID = r.Receiver_ID
GROUP BY r.Receiver_ID, r.Name
ORDER BY Total_Claims DESC;
"""

result_df = pd.read_sql(query, conn)
result_df
```

# Query 6

```python
# Which city has the highest number of food listings

query2_top = """
SELECT TOP 1 WITH TIES
    p.City,
    COUNT(f.Food_ID) AS Total_Listings
FROM Food_Listings_Dataset f
JOIN Providers p
    ON f.Provider_ID = p.Provider_ID
GROUP BY p.City
ORDER BY Total_Listings DESC;
"""
top_city_df = pd.read_sql(query2_top, conn)
display(top_city_df)
```

```python
# Question 8: How many food claims have been made for each food item?


query_8 = """
SELECT 
    fl.Food_Name,
    COUNT(c.Claim_ID) AS TotalClaims
FROM Food_Listings_Dataset fl
LEFT JOIN Claims c ON fl.Provider_ID = c.Food_ID
GROUP BY fl.Food_Name
ORDER BY TotalClaims DESC;
"""
df_q8 = pd.read_sql(query_8, conn)
print("Q8: Food claims per food item")
display(df_q8)
```

# Query 9

```python

# Question 9: Which provider has had the highest number of successful food claims?


query_9 = """
SELECT 
    p.Name AS ProviderName,
    COUNT(c.Claim_ID) AS SuccessfulClaims
FROM Providers p
JOIN Food_Listings_Dataset fl ON p.Provider_ID = fl.Provider_ID
JOIN Claims c ON fl.Provider_ID = c.Food_ID
WHERE c.Status = 'Completed'
GROUP BY p.Name
ORDER BY SuccessfulClaims DESC
OFFSET 0 ROWS FETCH NEXT 1 ROWS ONLY;
"""
df_q9 = pd.read_sql(query_9, conn)
print("Q9: Provider with highest number of successful food claims")
display(df_q9)
```

# Query 10


```python
# Question 10: What percentage of food claims are completed vs. pending vs. canceled?


query_10 = """
SELECT 
    Status,
    COUNT(*) AS Count,
    ROUND((COUNT(*) * 100.0 / (SELECT COUNT(*) FROM Claims)), 2) AS Percentage
FROM Claims
GROUP BY Status;
"""
df_q10 = pd.read_sql(query_10, conn)
print("Q10: Percentage of claims by status")
display(df_q10)
```

# Query 11

```python

# 11. Average quantity of food claimed per receiver

query_avg_quantity = """
SELECT 
    r.Receiver_ID,
    r.Name AS Receiver_Name,
    AVG(f.Quantity) AS Avg_Quantity_Claimed
FROM Claims c
JOIN Food_Listings_Dataset f
    ON c.Food_ID = f.Food_ID
JOIN Receivers r
    ON c.Receiver_ID = r.Receiver_ID
WHERE c.Status = 'Completed'
GROUP BY r.Receiver_ID, r.Name
"""

avg_quantity_df = pd.read_sql(query_avg_quantity, conn)
print("Average quantity of food claimed per receiver:")
display(avg_quantity_df)
```

# Query 12

```python

# 12. Meal type claimed the most

query_top_meal_type = """
SELECT 
    f.Meal_Type,
    COUNT(*) AS Claim_Count
FROM Claims c
JOIN Food_Listings_Dataset f
    ON c.Food_ID = f.Food_ID
WHERE c.Status = 'Completed'
GROUP BY f.Meal_Type
ORDER BY Claim_Count DESC
OFFSET 0 ROWS FETCH NEXT 1 ROWS ONLY;
"""

top_meal_df = pd.read_sql(query_top_meal_type, conn)
print("Most claimed meal type:")
display(top_meal_df)
```

# Query 13

```python
# 13. Total quantity of food donated by each provider

query_total_donations = """
SELECT 
    p.Provider_ID,
    p.Name AS Provider_Name,
    SUM(f.Quantity) AS Total_Quantity_Donated
FROM Food_Listings_Dataset f
JOIN Providers p
    ON f.Provider_ID = p.Provider_ID
GROUP BY p.Provider_ID, p.Name
ORDER BY Total_Quantity_Donated DESC
"""

total_donations_df = pd.read_sql(query_total_donations, conn)
print("Total quantity donated by each provider:")
display(total_donations_df)
```
